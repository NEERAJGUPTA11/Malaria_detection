```python
from keras.layers import Input, Lambda, Dense, Flatten
from keras.models import Model
```

```python
import zipfile
zip_ref=zipfile.ZipFile('/content/Dataset_2.zip','r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
from keras.applications.vgg19 import VGG19
from keras.applications.vgg16 import preprocess_input
from keras.models import Sequential
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
IMAGE_SIZE = [224, 224]
```

```python
train_path = '/content/Dataset/Train'
valid_path = '/content/Dataset/Test'
```

```python
vgg = VGG19(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
vgg.summary()
```

```python
for layer in vgg.layers:
  layer.trainable = False
```

```python
folders = glob('Dataset/Train/*')
```

```python
folders
```

```python
x = Flatten()(vgg.output)
```

```python
prediction = Dense(len(folders), activation='softmax')(x)

# create a model object
model = Model(inputs=vgg.input, outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

```python

test_datagen = ImageDataGenerator(rescale = 1./255)

training_set = train_datagen.flow_from_directory('Dataset/Train',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')

test_set = test_datagen.flow_from_directory('Dataset/Test',
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
r = model.fit(
  training_set,
  validation_data=test_set,
  epochs=125,
  steps_per_epoch=training_set.n // training_set.batch_size,
  validation_steps=test_set.n // test_set.batch_size
)
```

```python
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')
```

```python
plt.plot(r.history['accuracy'], label='train accuracy')  # Corrected key
plt.plot(r.history['val_accuracy'], label='validation accuracy')  # Corrected key
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
```
